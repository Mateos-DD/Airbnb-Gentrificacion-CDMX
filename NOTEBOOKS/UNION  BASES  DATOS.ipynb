```python
import pandas as pd

# Diccionario: {fecha_del_scrape: ruta_al_calendar}
cal_paths = {

    "2024-09-25": r"/content/drive/MyDrive/GENTRIFICACION 2025/DATOS AIRBNB /25 SEP 24/calendar.csv.gz",
    "2024-12-27": r"/content/drive/MyDrive/GENTRIFICACION 2025/DATOS AIRBNB /27 DIC 24/calendar.csv.gz",
    "2025-03-19": r"/content/drive/MyDrive/GENTRIFICACION 2025/DATOS AIRBNB /19 MARZ 25/calendar.csv.gz",
    "2025-06-25": r"/content/drive/MyDrive/GENTRIFICACION 2025/DATOS AIRBNB /25 JUN 2025/calendar.csv.gz",
}

cal_list = []

for snap_str, path in cal_paths.items():
    df = pd.read_csv(path, compression="infer")

    # Fecha del scrape (la misma para todo ese archivo)
    df["snapshot"] = pd.to_datetime(snap_str)

    # Limpieza bÃ¡sica consistente
    df["date"] = pd.to_datetime(df["date"])
    df["price_calendar"] = (
        df["price"]
          .astype(str)
          .str.replace("$", "", regex=False)
          .str.replace(",", "", regex=False)
          .str.strip()
    )
    df["price_calendar"] = pd.to_numeric(df["price_calendar"], errors="coerce")

    cal_list.append(df)

# Un solo DataFrame con todos los calendars
cal_all = pd.concat(cal_list, ignore_index=True)
```

```python
# Ordenamos por listing, fecha de estancia y snapshot (mÃ¡s reciente al final)
cal_all = cal_all.sort_values(["listing_id", "date", "snapshot"])

# Nos quedamos con la observaciÃ³n mÃ¡s reciente para cada listing_id + date
#cal_nodup = cal_all.drop_duplicates(subset=["listing_id", "date"], keep="last")
cal_all.drop_duplicates(subset=["listing_id", "date"], keep="last")
```

```python
del df
```

```python
# Si quieres considerar solo dÃ­as disponibles:
#cal_use = cal_nodup[cal_nodup["available"] == "t"].copy()

# Precio medio por noche en la ciudad (o en tu recorte espacial) por dÃ­a de estancia
ts_daily = (
    cal_all.groupby("date")["price_calendar"]
           .mean()
           .sort_index()
)

# Si quieres un DataFrame en vez de Serie:
#ts_daily = ts_daily.to_frame(name="avg_price_city")
```

```python
ts_daily
ts_daily.to_csv("serie_diaria.csv", encoding="utf-8")
```

```python
# Asegurar Ã­ndice de fecha
ts_daily.index = pd.to_datetime(ts_daily.index)

# Semanal (por semana calendario)
ts_weekly = ts_daily.resample("W").mean()

# Mensual
ts_monthly = ts_daily.resample("M").mean()
```

```python
list_paths = {

  "2024-09-25": r"/content/drive/MyDrive/GENTRIFICACION 2025/DATOS AIRBNB /25 SEP 24/listings.csv.gz",
    "2024-12-27": r"/content/drive/MyDrive/GENTRIFICACION 2025/DATOS AIRBNB /27 DIC 24/listings.csv.gz",
    "2025-03-19": r"/content/drive/MyDrive/GENTRIFICACION 2025/DATOS AIRBNB /19 MARZ 25/listings.csv.gz",
    "2025-06-25": r"/content/drive/MyDrive/GENTRIFICACION 2025/DATOS AIRBNB /25 JUN 2025/listings.csv.gz",
}

lis_list = []

for snap_str, path in list_paths.items():
    df = pd.read_csv(path, compression="infer")
    df["snapshot"] = pd.to_datetime(snap_str)

    # Renombrar id para que coincida con listing_id
    df = df.rename(columns={"id": "listing_id"})

    # Precio base del listing (opcional)
    df["price_listing"] = (
        df["price"]
          .astype(str)
          .str.replace("$", "", regex=False)
          .str.replace(",", "", regex=False)
          .str.strip()
    )
    df["price_listing"] = pd.to_numeric(df["price_listing"], errors="coerce")

    lis_list.append(df)

lis_all = pd.concat(lis_list, ignore_index=True)
```
