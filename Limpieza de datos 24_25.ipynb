```python
import numpy as np
import pandas as pd

def parse_bathrooms_text(s):
    s = s.astype(str).str.lower().str.strip()
    out = pd.Series(np.nan, index=s.index, dtype=float)
    # “half bath” sin número
    out[s.str.contains("half") & ~s.str.contains(r"\d")] = 0.5
    num = pd.to_numeric(s.str.extract(r"(\d+(\.\d+)?)")[0], errors="coerce")
    return out.fillna(num)

df_baths["bathrooms_parsed"] = parse_bathrooms_text(df_baths["bathrooms_text"])
```

```python
# La idea de este algoritmo es que a partir de la base de datos original "listings_gz_JUN", el "host_id"
#se repite es mas de una fila, por lo que podriamos buscar si en alguna de estas filas el valor en la columna
#"host_since" es no nulo, para rellenar esos 914 valores nulos.


df = listings_gz_JUN.copy()
df['host_id'] = df['host_id'].astype('string')  # pasamos a como año nada mas

#  Medimos faltantes o nulos
mask_null_orig = df['host_since'].isna()
n_missing_orig = int(mask_null_orig.sum())


donor_map = (
    df.dropna(subset=['host_since'])
      .groupby('host_id', as_index=True)['host_since']
      .min()   # año más antiguo observado para ese host
)


# Calcular imputaciones solo para las filas que estaban nulas
imputaciones = (
    df.loc[mask_null_orig, ['host_id']]
      .assign(imputed_year = lambda d: d['host_id'].map(donor_map))
      .dropna(subset=['imputed_year'])
)

imputaciones = imputaciones.sort_values(['imputed_year','host_id']).reset_index(drop=True)

host_since_filled = df['host_since'].copy()
host_since_filled = host_since_filled.fillna(
    pd.to_numeric(df['host_id'].map(donor_map), errors='coerce')
).astype('Int64')

df['host_since_filled'] = host_since_filled

#  Resumen
n_recovered = len(imputaciones)          # cuántas filas nulas recibieron valor
n_still_null = int(df['host_since_filled'].isna().sum())

print(f"Faltantes originales: {n_missing_orig}")
print(f"Recuperados con otros anuncios del mismo host: {n_recovered}")
print(f"Persisten nulos tras imputar: {n_still_null}")

print('\nImputaciones (host_id -> año):')
print(imputaciones)


anios_imputados_unicos = sorted(imputaciones['imputed_year'].unique().tolist())
print('\nAños imputados únicos:', anios_imputados_unicos)
```

```python
import pandas as pd


listings_gz_JUN = listings_gz_JUN.copy()
listings_gz_JUN['host_id'] = listings_gz_JUN['host_id'].astype(str)
listings_gz_JUN['host_since'] = pd.to_datetime(listings_gz_JUN['host_since'], errors='coerce')

# 1) host_id de filas con host_since nulo
mask_null = listings_gz_JUN['host_since'].isna()
host_ids_null = listings_gz_JUN.loc[mask_null, 'host_id'].dropna().unique()

# 2) ¿Se repiten con host_since NO nulo en otras filas de JUN?
repetidos_no_nulos_JUN = listings_gz_JUN.loc[
    (~listings_gz_JUN['host_since'].isna()) &
    (listings_gz_JUN['host_id'].isin(host_ids_null)),
    ['host_id', 'host_since']
].sort_values(['host_id','host_since'])

# 3) Resumen por host_id (cuántas veces aparecen nulos/no nulos en JUN)
resumen_JUN = (
    listings_gz_JUN
      .assign(is_null=listings_gz_JUN['host_since'].isna())
      .groupby('host_id')['is_null']
      .value_counts()
      .unstack(fill_value=0)
      .rename(columns={False:'non_null_rows_JUN', True:'null_rows_JUN'})
      .loc[host_ids_null]              # solo los host_id que tenían algún nulo
      .sort_values(['non_null_rows_JUN','null_rows_JUN'], ascending=False)
)
```
